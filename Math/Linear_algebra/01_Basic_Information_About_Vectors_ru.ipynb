```python
import matplotlib.pyplot as plt
import numpy as np

def plot_vectors(vectors):
    """
    Plot vectors on a 2D plane.

    Parameters:
    - vectors: List of tuples, where each tuple represents the components of a vector.
    """

    fig, ax = plt.subplots()
    ax.set_aspect('equal', adjustable='box')

    for vector in vectors:
        ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='blue', label=f"Vector {vector}")

    ax.axhline(y=0, color='black', linestyle='--')
    ax.axvline(x=0, color='black', linestyle='--')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Vectors in 2D Plane')
    plt.legend()
    plt.grid(True)
    plt.show()

# example:
vectors_to_plot = [(4, 2), (-2, 4), (2, -1)]
plot_vectors(vectors_to_plot)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vectors_with_scalar_product(a, b):
    scalar_product = np.dot(a, b)
    
    fig, ax = plt.subplots()
    ax.set_aspect('equal', adjustable='box')

    # Plot vectors
    ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
    ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector b')

    # Plot scalar product
    ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', alpha=0.5, linestyle='dashed')
    ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', alpha=0.5, linestyle='dashed')
    ax.quiver(b[0], b[1], a[0], a[1], angles='xy', scale_units='xy', scale=1, color='green', linestyle='dashed', label='Scalar Product (a  b)')

    ax.text(a[0]/2, a[1]/2, f'({a[0]}, {a[1]})', ha='right', va='top', color='blue')
    ax.text(b[0]/2, b[1]/2, f'({b[0]}, {b[1]})', ha='right', va='top', color='red')
    ax.text((a[0]+b[0])/2, (a[1]+b[1])/2, f'{scalar_product}', ha='right', va='top', color='green')

    ax.axhline(y=0, color='black', linestyle='--')
    ax.axvline(x=0, color='black', linestyle='--')

    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Scalar Product of Vectors')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example:
vector_a = np.array([2, 3])
vector_b = np.array([-1, 2])

plot_vectors_with_scalar_product(vector_a, vector_b)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_vectors_with_vector_product(a, b):
    vector_product = np.cross(a, b)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot vectors
    ax.quiver(0, 0, 0, a[0], a[1], a[2], color='blue', label='Vector a')
    ax.quiver(0, 0, 0, b[0], b[1], b[2], color='red', label='Vector b')

    # Plot vector product
    ax.quiver(0, 0, 0, vector_product[0], vector_product[1], vector_product[2], color='green', label='Vector Product (a x b)')

    ax.text(a[0], a[1], a[2], f'({a[0]}, {a[1]}, {a[2]})', color='blue')
    ax.text(b[0], b[1], b[2], f'({b[0]}, {b[1]}, {b[2]})', color='red')
    ax.text(vector_product[0], vector_product[1], vector_product[2], f'({vector_product[0]}, {vector_product[1]}, {vector_product[2]})', color='green')

    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('Vector Product of Vectors')
    ax.legend()

    plt.show()

# Example:
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

plot_vectors_with_vector_product(vector_a, vector_b)
```
